```python
# To summarize: 
review_rating_list = [a.attrs['data-rating'] for a in soup.select("article.profile-review-card div.rated-review--stars div.ng-isolate-scope")]
review_recipe_id_list = [a.attrs['data-ng-href'].split('/')[2] for a in soup.select("article.profile-review-card div.rated-review--stars a.ng-scope")]
review_recipe_links_list = [a.attrs['data-ng-href'] for a in soup.select("article.profile-review-card div.rated-review--stars a.ng-scope")]
review_recipe_title_list = [a.text for a in soup.select("article.profile-review-card h3.ng-binding")]
review_text_list = [a.text for a in soup.select("article.profile-review-card div.rated-review-text")]
```

```python
reviews = zip(review_recipe_id, review_recipe_title_list, review_rating_list, review_recipe_links, review_text_list)
```

```python
# For each review, create a dict w/ keys= ['recipe_ID', 'recipe_title', 'recipe_rating', 'recipe_link', 'review_text']
reviews_dict_list = [dict(zip(['recipe_ID', 'recipe_title', 'recipe_rating','recipe_link', 'review_text'], a)) for a in reviews]
```
