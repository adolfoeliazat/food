## 'Reviews' page in allrecipes.com's member page

Need to get: 
- recipe ID, name 
- rating for recipe 

- review text 


```python
import pymongo 
from bs4 import BeautifulSoup as BS
```

```python
client = pymongo.MongoClient('mongodb://localhost:27017/')
```

```python
allrecipes_db = client.allrecipes
```

```python
allrecipes_db.collection_names()
```

```python
members_coll = allrecipes_db.members
```

```python
members_coll
```

```python
# Get all the key names 
members_coll.find_one().keys()
```

```python
soup = BS(members_coll.find_one()['reviews_page'], 'html.parser')
```

```python
print soup.prettify()
```

### Find all the recipe links in 'reviews'.  


```python
# Find all the reviews'.  
```

```python
# This has all the recipes reviewed (but only on the first page), plus personal recipes and .  
# 
[a['href'] for a in soup.select('a.ng-scope')]
```

```python
set([a['data-ng-href'].split('/')[2] for a in soup.select("h3.grid-col__h3 a.ng-scope")])
```

```python
set([a.attrs['data-ng-href'] for a in soup.select("h3.grid-col__h3 a.ng-scope")])
```

```python
# Get recipe name 
set([a['href'].split('/')[3] for a in soup.select('a.ng-scope')])
```

```python
# Get recipe id 
set([a['href'].split('/')[2] for a in soup.select('a.ng-scope')])
```

```python
# These are the links 
set([a.attrs['href'] for a in soup.select('article.profile-review-card a.ng-scope')])
```

```python
# This is the review score: 
[a.attrs['data-rating'] for a in soup.select("article.profile-review-card div.rated-review--stars div.ng-isolate-scope")]
```

```python
# These are the reviewed recipes' links
[a.attrs['data-ng-href'] for a in soup.select("article.profile-review-card div.rated-review--stars a.ng-scope")]
```

```python
# These are the texts of the reviews 
[a.text for a in soup.select("article.profile-review-card div.rated-review-text")]
```

```python
# Can I get recipe name from within 'article.profile-review-card', for consistency's sake? Yes.  
[a.text for a in soup.select("article.profile-review-card h3.ng-binding")]
```

```python
[a.attrs['data-ng-href'].split('/')[2] for a in soup.select("article.profile-review-card div.rated-review--stars a.ng-scope")]
```

```python
# To summarize: 
review_rating_list = [a.attrs['data-rating'] for a in soup.select("article.profile-review-card div.rated-review--stars div.ng-isolate-scope")]
review_recipe_id_list = [a.attrs['data-ng-href'].split('/')[2] for a in soup.select("article.profile-review-card div.rated-review--stars a.ng-scope")]
review_recipe_links_list = [a.attrs['data-ng-href'] for a in soup.select("article.profile-review-card div.rated-review--stars a.ng-scope")]
review_recipe_title_list = [a.text for a in soup.select("article.profile-review-card h3.ng-binding")]
review_text_list = [a.text for a in soup.select("article.profile-review-card div.rated-review-text")]
```
