```python
paired = []
def pair_id_text():
    for i,p in enumerate(text_list): 
        if id_list[i] == []: 
            id = ""
        else: 
            id = id_list[i][0]
        if p == []: 
            t = ""
        else: 
            t = p[0]
        paired.append((id, t))
    
pair_id_text()
```

```python
for a in paired: 
    members_coll.find_one_and_update({"member_ID": a[0]}, 
                                    {"$set": {"aboutme": a[1]}}, 
                                    upsert = True)
    print "members with about me page:",members_coll.find({"aboutme": {"$exists": True}}).count()
```
