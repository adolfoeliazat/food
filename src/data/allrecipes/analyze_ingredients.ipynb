### NLP strategy for `ingredients`: 
* Structure of `ingredients_list`:
  * item on list starts with quantity (decimal point number or fraction) 
  * unit (e.g. 'cups', 'pinch', 'tablespoon', etc).  
  * ingredient noun 
  * Optionally followed by preparation ('diced, 'melted', 'softened') 
  * And/or optionally followed by descriptor (e.g. 'optional') 
  
See how well the sample dataset follows
